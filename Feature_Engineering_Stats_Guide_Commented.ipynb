## 1Ô∏è‚É£ Problem Statement
Hum check karna chahte hain:
**Kya categorical variable (Gender) ka numeric variable (Age) pe effect hai?**

2 cheeze use karenge:
- ANOVA ‚Üí Difference statistically significant hai ya nahi
- Eta Squared ‚Üí Difference kitna strong hai

## 2Ô∏è‚É£ ANOVA Test (Difference hai ya nahi?)

```
# ANOVA function import kar rahe hain scipy se
from scipy.stats import f_oneway

# Male rows filter kar rahe hain aur sirf Age column le rahe hain
male_age = df[df['Gender']=='Male']['Age']

# Female rows filter kar rahe hain aur sirf Age column le rahe hain
female_age = df[df['Gender']=='Female']['Age']

# Other rows filter kar rahe hain aur sirf Age column le rahe hain
other_age = df[df['Gender']=='Other']['Age']

# Teeno groups ke mean compare kar rahe hain
# Output me F-statistic aur p-value milegi
f_oneway(male_age, female_age, other_age)
```

## 3Ô∏è‚É£ Eta Squared (Effect Size - Kitna Strong Effect Hai?)

```
# Sab logon ka overall average age nikal rahe hain (Grand Mean)
grand_mean = df['Age'].mean()

# Gender ke hisaab se Age ko group kar rahe hain
groups = df.groupby('Gender')['Age']

# Har gender ka average age nikal rahe hain
group_means = groups.mean()

# Har gender me kitne log hain wo count kar rahe hain
group_sizes = groups.count()

# SS_between nikal rahe hain
# Ye measure karta hai groups ka mean overall mean se kitna door hai
# Difference ko square karte hain aur group size se multiply karte hain
ss_between = sum(group_sizes * (group_means - grand_mean)**2)

# SS_total nikal rahe hain
# Ye total dataset me kitni variation hai wo batata hai
ss_total = sum((df['Age'] - grand_mean)**2)

# Eta Squared formula apply kar rahe hain
# Total variation ka kitna percent Gender explain kar raha hai
eta_squared = ss_between / ss_total

# Final effect size print kar rahe hain
print('Eta Squared:', eta_squared)
```
