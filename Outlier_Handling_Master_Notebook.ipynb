## 1️⃣ Sample Data Create Karna

```python
# Basic libraries import kar rahe hain
import numpy as np
import pandas as pd

# Random seed set kar rahe hain taaki results reproducible rahe
np.random.seed(42)

# Normal distributed data create kar rahe hain
data = np.random.normal(50, 10, 100)

# Intentionally extreme values add kar rahe hain (outliers)
data = np.append(data, [150, 170, 200])

# DataFrame bana rahe hain
df = pd.DataFrame({'feature': data})

df.head()
```

## 2️⃣ IQR Method (Most Practical & Interview Favourite)

```python
# Q1 (25th percentile) calculate kar rahe hain
Q1 = df['feature'].quantile(0.25)

# Q3 (75th percentile) calculate kar rahe hain
Q3 = df['feature'].quantile(0.75)

# IQR calculate kar rahe hain
IQR = Q3 - Q1

# Lower aur Upper bound define kar rahe hain
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Outliers detect kar rahe hain
outliers_iqr = df[(df['feature'] < lower) | (df['feature'] > upper)]

outliers_iqr
```

## 3️⃣ Z-Score Method (Normal Distribution Ke Liye)

```python
# Mean calculate kar rahe hain
mean = df['feature'].mean()

# Standard deviation calculate kar rahe hain
std = df['feature'].std()

# Z-score calculate kar rahe hain
df['z_score'] = (df['feature'] - mean) / std

# |Z| > 3 ko outlier consider kar rahe hain
outliers_z = df[np.abs(df['z_score']) > 3]

outliers_z
```

## 4️⃣ Capping / Winsorization (Remove nahi karna, limit karna)

```python
# Lower aur Upper bound ke according clipping kar rahe hain
df['capped_feature'] = np.clip(df['feature'], lower, upper)

df.head()
```

## 5️⃣ Log Transformation (Skewed Data Ke Liye)

```python
# Log1p use kar rahe hain taaki zero values pe error na aaye
df['log_feature'] = np.log1p(df['feature'])

df.head()
```

## 6️⃣ Robust Scaling (Outlier Resistant Scaling)

```python
# RobustScaler import kar rahe hain
from sklearn.preprocessing import RobustScaler

# Object create kar rahe hain
scaler = RobustScaler()

# Scaling apply kar rahe hain
df['robust_scaled'] = scaler.fit_transform(df[['feature']])

df.head()
```

## 7️⃣ Model Based Handling (Tree Models Robust Hote Hain)
Linear regression outliers se sensitive hota hai.
Tree-based models jaise RandomForest usually robust hote hain.


```python
# Linear Regression aur RandomForest compare karne ke liye example
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Dummy target create kar rahe hain
df['target'] = df['feature'] * 2 + np.random.normal(0, 5, len(df))

# Train-test split kar rahe hain
X_train, X_test, y_train, y_test = train_test_split(df[['feature']], df['target'], test_size=0.2, random_state=42)

# Linear Regression train kar rahe hain
lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

# Random Forest train kar rahe hain
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

# R2 score compare kar rahe hain
print('Linear Regression R2:', r2_score(y_test, pred_lr))
print('Random Forest R2:', r2_score(y_test, pred_rf))
```

## 🎯 Final Notes (Interview Ready)
- IQR → Skewed data ke liye best
- Z-score → Normal data ke liye
- Capping → Data loss avoid karne ke liye
- Log transform → Right skew reduce karne ke liye
- RobustScaler → Median based scaling
- Tree models → Outlier resistant
