````python
import pandas as pd
````

````python
data = {
    'names': ['john','devid','jack','ruby','rahul','joe','grace','lana','max','pinky','eva'],
    'gender': ['male','male','male','female','male','female','female','female','male','female','female'],
    'cabin': ['B1-1000','z9-1299','y6-2304','a6-1223','g7-2356','q1-2165','t1-3454','h5-1278','p1-3267','z9-3489','l1-9090'],
    'city':['tokyo','helseinki','london','newyork','melbourne','sydney','paris','toronto','dubai','mumbai','berlin'],
    'age':[29,28,29,30,31,26,23,25,32,33,27]
}
````

````python
df = pd.DataFrame(data)
````

iss code ka use tab kiya jata hai jab column main se agar  ***alphabets***  aur ***numbers*** nikalna ho to. iss ke liye new column jo hum create karna chahte hai uss ke liye columns ko as a list main dena hota hai jaise yaha `df[['cabin_serial', 'cabin_number']]` iss code main  diya hai. iss main jo last wala part hai `(r'([A-Za-z0-9]+)-(\d+)')`  ye numbers aur alfabets ko extract karta hia. `([A-Za-z0-9]+)`  ye wala part alfabets ko extract kart hai aur ,`(\d+) `ye jo part ye numbers ko extract karne ke liye use hota hia.

````python
df[['cabin_serial', 'cabin_number']] = df['cabin'].str.extract(r'([A-Za-z0-9]+)-(\d+)')
````

ye code same vaise hi kam karta hia per iss mai pehle split hoti hia fiir list main convert hone ke bad list ke `[0]` iss ka use kar ke index **o** ko nikala.

````python
df['cabin_num'] = df['cabin'].str.split('-').str[0]
````

ye code bhi same hia jsie uper ka hia sirf iss main `[1]` iss ka use kar ke index **1** ko liya gya hia.

````python
df['cabin_alfa'] = df['cabin'].str.split('-').str[1]
````

# Colab Notebook ko GitHub Repository par Push Kaise Karein

Iss notebook ko `https://github.com/sachinmasti/future-engendering-techniques-/tree/sunfower` repository ke `sunfower` branch par push karne ke liye neeche diye gaye steps ko follow karein:

## Step 1: Apne Colab Notebook ko Save Karein

Sabse pehle, yeh ensure karein ki aapki saari latest changes save ho gayi hain. Aap `File` > `Save` par click kar sakte hain ya `Ctrl + S` (Windows/Linux) / `Cmd + S` (Mac) use kar sakte hain.

## Step 2: GitHub Personal Access Token (PAT) Generate Karein (Agar nahi kiya hai)

GitHub par changes push karne ke liye aapko Personal Access Token (PAT) ki zaroorat padegi. Agar aapke paas PAT nahi hai, toh neeche diye gaye steps follow karein:

1.  GitHub par login karein.
2.  Apni profile picture par click karein (top-right corner) aur `Settings` par jayein.
3.  Left sidebar mein `Developer settings` par click karein.
4.  `Personal access tokens` > `Tokens (classic)` par click karein.
5.  `Generate new token (classic)` par click karein.
6.  `Note` field mein apne token ka naam dein (jaise `Colab-Access`).
7.  `Expiration` set karein (jaise 30 din, 90 din, ya Custom).
8.  `Select scopes` mein `repo` checkbox ko tick karein (yeh repository access ke liye zaroori hai).
9.  Sabse neeche `Generate token` par click karein.
10. Token generate hone ke baad, use **turant copy karein** aur kisi safe jagah store karein. Yeh token aapko sirf ek baar dikhega.

## Step 3: Git Repository ko Clone Karein Colab mein

    ```bash
    !git clone https://YOUR_GITHUB_USERNAME:YOUR_PERSONAL_ACCESS_TOKEN@github.com/sachinmasti/future-engendering-techniques-.git
    ```
    *Agar aap token ko code mein expose nahi karna chahte hain, toh aap `git clone https://github.com/sachinmasti/future-engendering-techniques-.git` run kar sakte hain. Jab credentials maange jayenge, tab username aur PAT enter karein.*
