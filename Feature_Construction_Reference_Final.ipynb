## 📦 Import Libraries

👉 Pandas use karenge data handle karne ke liye

```python
import pandas as pd
```

## 🗓️ 1. Date Feature Construction

👉 Problem: Model ko raw date samajh nahi aata
👉 Solution: Date ko tod ke useful parts banao

Kab use kare?
- jab time-based pattern ho (sales, user activity, etc.)

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek
df['is_weekend'] = df['dayofweek'].isin([5,6]).astype(int)
```

## 🔢 2. Count-Based Features

👉 Problem: Multi-value columns (comma separated)
👉 Solution: split karo aur count nikaalo

Kab use kare?
- jab ek row me multiple values ho (cast, genres, tags)

```python
df['num_cast'] = df['cast'].fillna('').str.split(',').apply(len)
df['num_genres'] = df['listed_in'].fillna('').str.split(',').apply(len)
```

## 🌍 3. Multi-Value Feature Flags

👉 Problem: ek se zyada countries ho sakti hai
👉 Solution: simple flag bana do

Kab use kare?
- jab complexity reduce karni ho

```python
df['multi_country'] = df['country'].fillna('').str.split(',').apply(lambda x: 1 if len(x) > 1 else 0)
```

## 🎯 4. Binary Flags (Keyword Based)

👉 Problem: categories bahot zyada hai
👉 Solution: important keywords ke flags bana

Kab use kare?
- jab high cardinality ho

```python
df['is_drama'] = df['listed_in'].str.contains('Drama', na=False).astype(int)
df['is_comedy'] = df['listed_in'].str.contains('Comedy', na=False).astype(int)
```

## 📊 5. Aggregation Features

👉 Problem: individual row weak hoti hai
👉 Solution: group level behavior add karo

Kab use kare?
- user, product, category level data ho

```python
df['user_total'] = df.groupby('user_id')['amount'].transform('sum')
df['user_avg'] = df.groupby('user_id')['amount'].transform('mean')
df['user_count'] = df.groupby('user_id')['amount'].transform('count')
```

## 🔥 6. Interaction Features

👉 Problem: single feature weak ho sakta hai
👉 Solution: features combine karo

Kab use kare?
- jab 2 features milke better signal dete ho

```python
df['age_duration'] = df['content_age'] * df['duration']
df['cast_per_duration'] = df['num_cast'] / (df['duration'] + 1)
```

## 🧠 Final Rules (Yaad rakhna bro)

- Split → Count (multi-value ke liye)
- High dimension avoid kar
- Flags use kar instead of full encoding
- Always soch: 'model ko kya signal mil raha hai?'
