##iss niche ke cells main jo code hai wo date time data ke liye usefull hia.

sabse pehle jo `datetime` column hota hai wo ek object hota hai uss ko pehle **pandas** ke `datetime` main convert karna padta.

```python
df['date'] = pd.to_datetime(df['date'])
```

iss main jo function's hia `dt` ke bad uss se hi pata cahlta hai ki wo function kaha use karna hia.
*for ex*   jo `dt.year` hai uss ka use year ko nikalne ke liye hota hia.`dt.month` ka use month ko nikalne ke liye hota hia.and so on.

```python
df['year']        = df['date'].dt.year
df['month']       = df['date'].dt.month
df['day']         = df['date'].dt.day
df['hour']        = df['date'].dt.hour
df['day_name']    = df['date'].dt.day_name()    # 'Monday'
df['month_name']  = df['date'].dt.month_name()  # 'January'
df['is_weekend']  = df['date'].dt.dayofweek.isin([5,6]).astype(int)
df['quarter']     = df['date'].dt.quarter
```

iss code ka use `date` ke difference nikalne ke liye hota hia.

```python
df['days_since'] = (pd.Timestamp.now() - df['date']).dt.days
```

iss niche ke cell man jo code hai uss ka use **date wise comparison ke liye hota hia**.

```python
# Ek specific month
df[df['date'].dt.month == 1]

# Date range
df[(df['date'] >= '2024-01-01') & (df['date'] <= '2024-12-31')]
```

ye jo code hia iss ka use `month` aur  `day` ke wise grouopby karne ke liye hota hai. iss code main **sales** ko month wise groupby kiya hia.aur **day** wise bhi kiya gya hia.

```python
# Month wise sales
df.groupby(df['date'].dt.month)['sales'].sum()

# Day wise
df.groupby(df['date'].dt.day_name())['sales'].mean()
```

iss code ka use common format specify karne ke liye dekhaya gya hia. **hum khud `datetime` ko format kar sakte hia jaise chahiye wise**.

```python
# Format specify karo agar auto detect na ho
pd.to_datetime(df['date'], format='%d-%m-%Y')  # 25-01-2024
pd.to_datetime(df['date'], format='%Y/%m/%d')  # 2024/01/25
```

<span style="color:pink;"> Pro Tip  DateTime ko convert karne ke baad index set karo time series data ke liye:</span>


```python
df.set_index('date', inplace=True)
df['2024']        # sirf 2024 ka data
df['2024-01']     # sirf January 2024
```
